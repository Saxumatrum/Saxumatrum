```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from imblearn.ensemble import BalancedRandomForestClassifier, EasyEnsembleClassifier
import xgboost as xgb
from catboost import CatBoostClassifier
```

```python
go=pd.read_csv('GoDataset_updated.csv')
```

```python
go['Class'] = go['Class'].replace('CR', 1)
go['Class'] = go['Class'].replace('NotCR', 0)
```

```python
#X = go.iloc[:,:-1]#select all except the last 2 columns
#y = go.iloc[:,-1:]#select the last column
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
#train_data=pd.concat([X_train,y_train],axis=1)
from sklearn.model_selection import cross_val_score

# Undersampling loop
num_iterations = 20  # Number of iterations for undersampling
undersampled_data = []  # List to store the undersampled training datasets
brf_models= []  # List to store the BRF models
brf_auc_accuracies=[]
brf_f1_accuracies=[]
brf_recall_accuracies=[]
count=0


for i in range(num_iterations):
    X = go.iloc[:,:-1]#select all except the last 2 columns
    y = go.iloc[:,-1:]#select the last column
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


    # Separate majority and minority classes
    #majority_class = train_data[train_data['Class'] == 0]
    #minority_class = train_data[train_data['Class'] == 1]

    # Undersample majority class to match the size of the minority class
    #majority_undersampled = majority_class.sample(len(minority_class), random_state=42)

    # Combine the undersampled majority class with the minority class
    #undersampled_data.append(pd.concat([majority_undersampled, minority_class], axis=0))

    # Access each undersampled training dataset and create a BRF model
   # X_train_undersampled = i.iloc[:, :-1]
    #y_train_undersampled = i['Class']

    # Fit the  model on the undersampled data
    brf=BalancedRandomForestClassifier(n_estimators=500, random_state=42)
    brf.fit(X_train,y_train)
    print(np.mean(cross_val_score(brf, X_train, y_train, cv=3)))

    test_accuracy = brf.score(X_test, y_test)
    print("Test Accuracy:", test_accuracy)

    #calulate f1 and recall
    y_pred = brf.predict(X_test)

    # Compute F1 score and recall
    f1 = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    y_pred_prob = brf.predict_proba(X_test)[:, 1]
# Calculate the AUC score
    auc_score_brf = roc_auc_score(y_test, y_pred_prob)


    brf_auc_accuracies.append(auc_score_brf)
    brf_f1_accuracies.append(f1)
    brf_recall_accuracies.append(recall)


    # Store the model in the list along with the undersampled dataset index
    brf_models.append((i, brf))
    count=count+1
    print(count)
```
